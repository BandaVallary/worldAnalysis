```python

# #Whenever Pandas does not recognize the data type as one of the small handful of datatypes it can deal with (int, float, string, boolean, …),
# it just sets the datatype to “object” — that’s a safe bet, since pretty much everything is an object, in Python.
# So, it can be anything. It could even be an int, or float, or string, if you wish. It could be a column full of functions. It could be instantiations of your own class.
#this means that pandas/python cannot automatically perform arithmetic operations

#for these columns we need to convert them to integers by removing the commas and then convert them to integers
        # world_df['Population'] = world_df['Population'].replace(',','', regex=True)
        # world_df['Urban_population'] = world_df['Urban_population'].replace(',','', regex=True)

world_df[['Population', 'Urban_population', 'Density\n(P/Km2)', 'Land Area(Km2)', 'Armed Forces size','Co2-Emissions']]=world_df[['Population', 'Urban_population', 'Density\n(P/Km2)', 'Land Area(Km2)', 'Armed Forces size', 'Co2-Emissions']].replace(',','', regex=True)

# #conversion to integers
world_df =world_df.astype({
          'Population':'Int64', 'Urban_population':'Int64', 'Density\n(P/Km2)':'Int64', 'Land Area(Km2)':'Int64', 'Armed Forces size':'Int64',
          'Co2-Emissions':'Int64'
        })
```

```python
#this column's values contains commas and decimals. Decimals are automatically parsed into floats. Ours is to remove the commas.

world_df['CPI']=world_df['CPI'].str.replace(',','').astype(float)
```

```python
#the following columns have the % sign together with the numbers, therefore cannot be converted to float as is.
#we need to remove the sign and then convert. There are several ways to do that.

world_df['Agricultural Land( %)'] = world_df['Agricultural Land( %)'].str.strip('%').astype(float)
world_df['CPI Change (%)'] =world_df['CPI Change (%)'].str.strip('%').astype(float)
world_df['Out of pocket health expenditure']=world_df['Out of pocket health expenditure'].str.strip('%').astype(float)
world_df['Forested Area (%)']=world_df['Forested Area (%)'].str.strip('%').astype(float)
world_df['Gross primary education enrollment (%)']=world_df['Gross primary education enrollment (%)'].str.strip('%').astype(float)
world_df['Gross tertiary education enrollment (%)']=world_df['Gross tertiary education enrollment (%)'].str.strip('%').astype(float)
world_df['Population: Labor force participation (%)']=world_df['Population: Labor force participation (%)'].str.strip('%').astype(float)
world_df['Tax revenue (%)']=world_df['Tax revenue (%)'].str.strip('%').astype(float)
world_df['Total tax rate']=world_df['Total tax rate'].str.strip('%').astype(float)
world_df['Unemployment rate']=world_df['Unemployment rate'].str.strip('%').astype(float)

world_df
```

```python
#these columns contains $ symbol and is of type object, we want to convert to floating point
world_df['Gasoline Price']=world_df['Gasoline Price'].str.strip('$').astype(float)
world_df['Minimum wage']=world_df['Minimum wage'].str.strip('$').astype(float)
```

```python
#strings of numbers separated by commas cannot be converted directly to a float.
#it throws a valueError of invalid literal for int() with base 10. Therefore, we firts remove the commas by using the replace() method, remove the $ symbol & then use the astype method to convert
world_df['GDP']= world_df['GDP'].str.replace(',', '').str.strip('$').astype(float)

```

```python
#feature engineering
world_df['GDP per Capita']= (world_df['GDP']/world_df['Population']).round(2)
pd.options.display.float_format = '{:.4f}'.format
```
